# MACHINE LEARNING

```python
#data data yang sifatnya numerikal
new_data.select_dtypes(include = 'number')
```

>untuk data id, secara komputer akan dilihat sebagai angka. namun secara behavior, hanya sebagai identifier.  

```python
#membuat function yang bisa diterapkan di data testing
def select_columns(df: pd.DataFrame, features: list, target: list = 'rating'): #dikasih default parameter
  X = df[features]
  y = (df["rating"] >= 3).astype(int) #diubah jadi int

  return X, y #untuk membagi data kita
```

```python
#import library

from sklearn.model_selection import train_test_split
```

```python
#mengambil features untuk belajar
features = ["budget", "popularity", "runtime", "vote_average"]
X, y = select_columns(new_data, features = features)
```

```python
y
```

```python
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
#random state gunanya untuk memastikan bahwa proses pengacakan
#antara satu komputer dengan yg lainnya itu menghasilkan hal yang sama
```

```python
x_test
```

# Scalling

```python
#import library

from sklearn.preprocessing import MinMaxScaler
```

```python
#scalling berguna untuk memastikan tidak ada bias dari sisi satuan datanya
minmax = MinMaxScaler()

minmax.fit(X_train)
```

```python
#dijadiin ke dataframe
# pd.DataFrame(X_test_scaled, columns = minmax.get_feature_names_out())

#dalam function

def scaling(df, scaler):
  scaled_array = scaler.transform(df)
  scaled_df = pd.DataFrame(scaled_array, columns = scaler.get_feature_names_out())

  return scaled_df
```

```python
#mengaplikasikan xtrainnya
X_train_scaled = scaling(X_train, minmax)
X_test_scaled = scaling(x_test, minmax)
```

# Supervised Machine Learning

```python
##KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train_scaled, y_train)
```

```python
#nyaritau model yang dipake bagus apa gak
y_train_pred = knn.predict(X_train_scaled)
```

```python
#nyaritau model yang dipake bagus apa gak
y_test_pred = knn.predict(X_test_scaled)
```

```python
pd.crosstab(y_test, y_test_pred) #metrik evaluasi
```

```python
## Decision Tree

from sklearn.tree import DecisionTreeClassifier
```

```python
dt.fit(X_train_scaled, y_train)
```

```python
knn_params = {
    "n_neighbors": [3, 5, 7, 9, 11],
    "p": [1, 2, 3, 4, 5]
}

```

```python
from sklearn.model_selection import GridSearchCV
```

```python
knn_new = GridSearchCV(KNeighborsClassifier(),
                       knn_params, scoring = f1_score)
```

```python
knn_new.fit(X_train_scaled, y_train)
```

```python
knn_new.best_params_
```

```python
knn_best = knn_new.best_estimator_
```

```python
knn_new.best_score_
```

```python
prob = knn_best.predict_proba(X_test_scaled)[:,1]
```

```python
(prob > 0.5).astype("int")
```

```python
knn_best.predict(X_test_scaled)
```

```python
recall_ = []
precision_ = []
f1_score_list = []
accuracy_ = []
for i in range(1,11):
  y_best_knn_proba = knn_best.predict_proba(X_test_scaled)[:,1]
  recall_.append(recall_score(y_test, y_best_knn_proba >= i/10))
  precision_.append(precision_score(y_test, y_best_knn_proba >= i/10))
  f1_score_list.append(f1_score(y_test, y_best_knn_proba >= i/10))
  accuracy_.append(accuracy_score(y_test, y_best_knn_proba >= i/10))
```

```python
metrics = pd.DataFrame({"recall": recall_,
             "precision": precision_,
             "f1_score": f1_score_list,
             "accuracy": accuracy_})
```

```python
metrics["fpr"] = 1 - metrics["recall"]
```

```python
metrics.plot(x = "fpr", y = "recall")
```
