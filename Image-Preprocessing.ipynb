```python
#Cropping
from PIL import Image
import numpy as np
import os
import glob
import cv2
from google.colab.patches import cv2_imshow

fp2='/content/drive/MyDrive/Data/filter'
files = [f for f in os.listdir(fp2) if f.endswith(('.jpg', '.png', '.jpeg', '.gif'))]
folseg='/content/drive/MyDrive/Data/cropping'
# Loop melalui setiap file gambar
for filename in files:
    # Baca gambar
    image_path = os.path.join(fp2, filename)
    image = cv2.imread(image_path)
    if image is not None:
      height, width, channels = image.shape #Menentukan lebar dan tinggi

      #menentukan titik tengah
      xcenter = width/2
      ycenter = height/2

      #menetukan koordinat cropping
      x1 = int(xcenter-100)
      y1 = int(ycenter-100)
      x2 = int(xcenter+100)
      y2 = int(ycenter+100)

      crop = image[y1:y2, x1:x2] #cropping image

      # Simpan gambar yang telah diolah
      processed_image_path = os.path.join(folseg,'crop_' + filename)
      cv2.imwrite(processed_image_path, crop)
    else:
      print(f'Gagal membaca gambar: {image_path}')
```

```python
import os
import cv2
import numpy as np
import pandas as pd
from PIL import Image

# Convert RGB to HSI
def rgb_to_hsi(r, g, b):
    # Normalize RGB components to the range [0, 1]
    r, g, b = r / 255.0, g / 255.0, b / 255.0

    # Calculate Hue (H)
    theta = np.arccos(0.5 * ((2 * r - g - b) / np.sqrt((r - g) ** 2 + (r - b) * (g - b))))
    h = theta if b <= g else 2 * np.pi - theta
    h = (h / (2 * np.pi)) * 360  # Convert radians to degrees

    # Calculate Saturation (S)
    minimum = min(r, g, b)
    s = 1 - (3 * minimum / (r + g + b)) if (r + g + b) != 0 else 0

    # Calculate Intensity (I)
    i = (r + g + b) / 3.0

    return h, s, i

fp4 = '/content/drive/MyDrive/Data/kilau'
files = [f for f in os.listdir(fp4) if f.endswith(('.jpg', '.png', '.jpeg', '.gif'))]
files = sorted(files)
filex = '/content/drive/MyDrive/Data'

# Initialize empty lists to store results
r_list = []
g_list = []
b_list = []
h_list = []
s_list = []
i_list = []
kls_list = []

# Loop through each image file
for i in files:
    # Read the image
    image_path = os.path.join(fp4, i)

    kelas=i.split('_')[-1]
    kelas=kelas.split('.')[0]
    kelas=kelas[:-2]
    if kelas=='mentah':
      kls=0
    elif kelas=='stenghmatang':
      kls=1
    elif kelas=='matang':
      kls=2
    elif kelas=='sgtmatang':
      kls=3
    print(i)
    a = np.array(Image.open(image_path))

    if a is not None:
        r = np.mean(a[:, :, 0])
        g = np.mean(a[:, :, 1])
        b = np.mean(a[:, :, 2])
        print(r)
        h, s, i = rgb_to_hsi(r, g, b)

        r_list.append(r)
        g_list.append(g)
        b_list.append(b)
        h_list.append(h)
        s_list.append(s)
        i_list.append(i)
        kls_list.append(kls)

    else:
        print(f'Failed to read the image: {image_path}')

# Create a DataFrame from the lists
RGBHSI = pd.DataFrame({'R': r_list, 'G': g_list, 'B': b_list, 'H': h_list, 'S': s_list, 'I': i_list, 'Kelas': kls_list})

# Save the processed data to an Excel file
output_path = os.path.join(filex, 'ekstraksihsi.xlsx')
RGBHSI.to_excel(output_path, index=False)
print(RGBHSI)
```
