# **Klasifikasi Tingkat Kematangan**

Berdasarkan warna HSI yang diperoleh dari citra buah jambu kristal dalam proses [pengolahan citra](https://colab.research.google.com/drive/1q99ChhSbiL0nG23IhTzrpeYgbQorxL9T#scrollTo=aeSgzzp2KXl-), tingkat kematangan buah dapat ditentukan menggunakan metode klasifikasi yang tepat. Metode jaringan saraf tiruan Backpropagation merupakan salah satu metode klasifikasi yang menyerupai cara kerja otak manusia dimana permasalahan diselesaikan dengan pembelajaran atau pelatihan terhadap jaringan. Jaringan saraf tiruan Backpropagation banyak digunakan untuk menyelesaikan permasalahan pengenalan pola, salah satunya yaitu klasifikasi


```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.utils import to_categorical

#memanggil data
data=pd.read_excel("/content/drive/MyDrive/Data/ekstraksihsi.xlsx")
data=pd.DataFrame(data,columns=['H','S','I','Kelas'])

#normalisasi H
scaler = MinMaxScaler()
H=np.array(data['H'])
H = H.reshape(-1, 1) #mengubah array 1D to 2D
scaled_data = scaler.fit_transform(H)
data['H']=scaled_data #mengganti nilai H

print(data)
#memisahkan data dan target
x_train=data.iloc[0:120,0:3].values #variabel x
train=data.iloc[0:120,3].values #variabel y
x_test=data.iloc[120:140,0:3].values #variabel x
test=data.iloc[120:140,3].values #variabel y
y_train = to_categorical(train)
y_test = to_categorical(test)
print(x_train)
print(y_train)
print(x_test)
print(y_test)
print((y_train.shape))
```

```python
import numpy as np
import tensorflow as tf #backpropagation
from tensorflow import keras

hidden=10
alpha=0.3
epochs=1500

# Membuat model arsitektur
model = tf.keras.models.Sequential()
# Inisialisasi bobot
weight1 = np.random.rand(3, hidden)
weight2 = np.random.rand(hidden, 4)
init1 = tf.constant_initializer(weight1)
init2 = tf.constant_initializer(weight2)
model.add(tf.keras.layers.Dense(units=hidden, activation='sigmoid', input_dim=3, kernel_initializer=init1)) #hidden layer
model.add(tf.keras.layers.Dense(units=4, activation='sigmoid', kernel_initializer=init2)) #output layer

opt = keras.optimizers.Adam(learning_rate=alpha)
model.compile(optimizer=opt,loss='categorical_crossentropy',metrics=['accuracy'])

print('training')
model.fit(x_train,y_train,epochs=epochs,batch_size=64)
```

```python
print('testing')
model.evaluate(x_test,y_test,batch_size=32)
```

```python
from sklearn import metrics

#Metrik evaluasi testing
pred = model.predict(x_test)
pred = np.argmax(pred, axis=1)
print(pred)
print(test)

average = 'weighted'
Accuracy = metrics.accuracy_score(test, pred)
Precision = metrics.precision_score(test, pred, average=average)
Recall = metrics.recall_score(test, pred, average=average)
F1_score = metrics.f1_score(test, pred, average=average)

#metrics
print({"Accuracy":Accuracy,"Precision":Precision,"Recall":Recall,"F1_score":F1_score})
```
